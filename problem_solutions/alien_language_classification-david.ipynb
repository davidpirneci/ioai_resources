```python
import numpy as np
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
from collections import Counter
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB, BernoulliNB, ComplementNB
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df_train_samples = pd.read_csv("train_samples.txt", sep="	", header=None, names=["id", "text"])
df_train_labels = pd.read_csv("train_labels.txt", sep="	", header=None, names=["id", "class"])
df_train_labels = df_train_labels.drop(columns=['id'])
df_train = pd.concat([df_train_samples, df_train_labels], axis=1)

df_train.head()
```

```python
df_test_samples = pd.read_csv("validation_samples.txt", sep="	", header=None, names=["id", "text"])
df_test_labels = pd.read_csv("validation_labels.txt", sep="	", header=None, names=["id", "class"])
df_test_labels = df_test_labels.drop(columns=['id'])
df_test = pd.concat([df_test_samples, df_test_labels], axis=1)

df_test.tail()
```

```python
text_list = df_train['text'].tolist()
class_list = df_train['class'].tolist()

# Preprocessing train data

count_vect = CountVectorizer(analyzer='char', ngram_range=(2,8), binary=True, encoding="UTF-8", lowercase=False, strip_accents="unicode")
X_train_counts = count_vect.fit_transform(text_list)

X_train = X_train_counts
y_train = class_list
```

```python
text_list1 = df_test['text'].tolist()
class_list1 = df_test['class'].tolist()

# Preprocessing test data

X_test_counts = count_vect.transform(text_list1)

X_test = X_test_counts
y_test = class_list1
```

```python
Model = MultinomialNB(alpha=0.2,fit_prior=True)
Model.fit(X_train, y_train)
y_pred = Model.predict(X_test)

print("Accuracy", accuracy_score(y_test, y_pred))
print("Macro precision_recall_fscore_support")
print(precision_recall_fscore_support(y_test, y_pred, average='macro'))
print("Micro precision_recall_fscore_support")
print(precision_recall_fscore_support(y_test, y_pred, average='micro'))
print("Weighted precision_recall_fscore_support")
print(precision_recall_fscore_support(y_test, y_pred, average='weighted'))
```

```python
Model = SVC(kernel='linear').fit(X_train, y_train)
y_pred = Model.predict(X_test)

print("Acuracy", accuracy_score(y_test, y_pred))
print("Macro precision_recall_fscore_support")
print(precision_recall_fscore_support(y_test, y_pred, average='macro'))
print("Micro precision_recall_fscore_support")
print(precision_recall_fscore_support(y_test, y_pred, average='micro'))
print("Weighted precision_recall_fscore_support")
print(precision_recall_fscore_support(y_test, y_pred, average='weighted'))
```

```python
Model = BernoulliNB()
Model.fit(X_train.toarray(), y_train)
y_pred = Model.predict(X_test.toarray())

print("Acuracy", accuracy_score(y_test, y_pred))
print("Macro precision_recall_fscore_support")
print(precision_recall_fscore_support(y_test, y_pred, average='macro'))
print("Micro precision_recall_fscore_support")
print(precision_recall_fscore_support(y_test, y_pred, average='micro'))
print("Weighted precision_recall_fscore_support")
print(precision_recall_fscore_support(y_test, y_pred, average='weighted'))
```

```python
Model = ComplementNB()
Model.fit(X_train.toarray(), y_train)
y_pred = Model.predict(X_test.toarray())

print("Acuracy", accuracy_score(y_test, y_pred))
print("Macro precision_recall_fscore_support")
print(precision_recall_fscore_support(y_test, y_pred, average='macro'))
print("Micro precision_recall_fscore_support")
print(precision_recall_fscore_support(y_test, y_pred, average='micro'))
print("Weighted precision_recall_fscore_support")
print(precision_recall_fscore_support(y_test, y_pred, average='weighted'))
```
